```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics # to check trained model accuracy
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree # visualizes decsion tree strcuture
```

```python
 # Set a random seed for reproducibility
 np.random.seed(42)
 # Generate data for class 0
 feature_1_class_0 = np.random.normal(2, 1, 100)
 feature_2_class_0 = np.random.normal(2, 1, 100)
 # Generate data for class 1
 feature_1_class_1 = np.random.normal(6, 1, 100)
 feature_2_class_1 = np.random.normal(5, 1, 100)
```

```python
 # combinign features and lables
 features_class_0 = np.column_stack((feature_1_class_0, feature_2_class_0)) #array of shape (100,2) and c
 features_class_1 = np.column_stack((feature_1_class_1, feature_2_class_1))
 #print (features_class_0)
 #print(features_class_1.shape) labels_class_0 = np.zeros(100)
 # albelling selected class with zero of the 100 values in it
 labels_class_0 = np.zeros(100)
 labels_class_1 = np.ones(100)
 features= np.concatenate((features_class_0, features_class_1))
 labels = np.concatenate((labels_class_0, labels_class_1))
 print(features)
 print(labels)
```

```python
 data = pd.DataFrame (features, columns = ['feature_1', 'feature_2'])
 data['Target'] = labels
 print(data)
```

```python
 x = data[['feature_1', 'feature_2']] # features
 y = data['Target'] # labels
 print(x)
 print(y)
```

```python
 x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
 clf = DecisionTreeClassifier (max_depth=4, random_state = 42)
 clf.fit(x_train, y_train)
 y_pred = clf.predict(x_test)
 accuracy = metrics.accuracy_score(y_test, y_pred)
 print("Accuracy:", accuracy)
```

```python
 plt.figure(figsize=(20, 12))
 plot_tree(clf, filled=True, feature_names=['feature_1', 'feature_2'], class_names=['Class 0', 'Class 1'])
 plt.show()
 #gini value refers to gini Imputity at that node
 #gini impurity is a measure of how pure or mised the chasses are within the node
```
