```python
data = pd.read_csv("supermarket_sales 2.csv")
```

```python
data['Unit price'].fillna(data['Unit price'].mean(), inplace = True)
data.head(15)
```

```python
data['Unit price'].fillna(data['Unit price'].median(), inplace = True)
data.head(15)
```

```python
data['Unit price'].fillna(data['Unit price'].mode()[0], inplace = True)
data.head(15)
```

```python
Q1 = data['Unit price'].quantile(0.25)
Q1
```

```python
Q3 = data['Unit price'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_bound = Q1
lower_bound
```

```python
upper_bound = Q3
upper_bound
```

```python
data = data[(data['Unit price'] > lower_bound) & (data['Unit price'] < upper_bound)]
data.head(12)
```

```python
data['Unit price'] = np.where(data['Unit price'] > upper_bound, upper_bound, data['Unit price'])
data.head(12)
```

```python
data['Unit price'] = np.where(data['Unit price'] < lower_bound, lower_bound, data['Unit price'])
data.head(12)
```

```python
data['Unit price'] = np.log(data['Unit price'])
data
```
