```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales.csv")
```

```python

ata = pd.read_csv("supermarket_sales.csv")
data.info()d
```

```python
#1. Line Plot (Matplotlib)
plt.figure()
plt.plot(data['Unit price'], data['Total'], marker='*')
plt.title('Line Plot')
plt.xlabel('Unit Price')
plt.ylabel('Total')
plt.show()
```

```python
#2. Bar Chart (Matplotlib)
plt.figure()
plt.bar(data['Quantity'],data['cogs'])
plt.title('Bar Chart')
plt.xlabel('Quantity')
plt.ylabel('cogs')
plt.show()
```

```python
# 3.scatter plot (Matplotlib)
plt.figure()
plt.scatter(data['Quantity'], data['Unit price'])
plt.title('Scatter Plot')
plt.xlabel('Quantity')
plt.ylabel('Unit price')
plt.show()
```

```python
# 4.Histogram (seaborn)
plt.figure()
sns.histplot(data['Quantity'], bins=6)
plt.title('histogram')
plt.xlabel('Ouantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#5. Box Plot(Seaborn)
plt.figure()
sns.boxplot(x='Quantity', y='Tax 5%', data=data)
plt.title('Box Plot')
plt.xlabel('Qunatity')
plt.ylabel('TaxÂ 5%')
plt.show()
```

```python
# 6.HeatMap (seaborn)
af = data[['Unit price', 'Quantity', 'Tax 5%', 'cogs', 'gross income', 'Rating']]
corr = af.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.show()
```
