```python
df.dropna(inplace=True)
```

```python
df.fillna(df['Unit price'].mean(), inplace=True)
```

```python
df.head(15)

```

```python
sns.histplot(df['Unit price'])
plt.show()
```

```python
sns.boxplot(x='Payment', y='Unit price', data=df)
plt.show()
```

```python
sns.scatterplot(x='cogs', y='Unit price', data=df)
plt.show()
```

```python
sns.countplot(x='Branch', data=df)
plt.show()
```

```python
af=df[['Unit price','Quantity','Tax 5%','cogs','gross income','Rating']]
corr = af.corr()
sns.heatmap(corr, annot=True)
plt.show()
```
